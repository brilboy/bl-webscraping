```python
#Import Library
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import os
```

```python
#Input API

key = input('What do you want to search?')
write = csv.writer(open('{}.csv'.format(key), 'w', newline=''))
header = ['Nama', 'Harga', 'Harga Diskon', 'Diskon', 'Stok', 'Terjual', 'Kota', 'Provinsi', 'Link']
write.writerow(header)

api = 'https://api.bukalapak.com/multistrategy-products'
count =0
for page in range(1,11):
  parameter = {
      'category_id': 375, #kode untuk Fullbike > Sepeda Lipat
      'keywords': key,
      'limit': 50,
      'offset': 50,
      'facet': True,
      'page': page,
      'shouldUseSeoMultistrategy': False,
      'isLoggedIn': False,
      'show_search_contexts': True,
      'access_token': 'wtGd8p1Gu0cbez4nEA67iK9Fa6aFOYU-bBAiYG2SVymsmQ'
  }
  #note: if error, please update the access_token

  #Input Variable
  req = requests.get(api, params=parameter).json()

  products = req['data']
  for p in products: #looping
    nama = p['name']
    harga = p['original_price']
    harga_diskon = p['price']
    besar_diskon = p['discount_percentage']
    stok = p['stock']
    terjual = p['stats']['sold_count']
    kota = p['store']['address']['city']
    provinsi = p['store']['address']['province']
    link = p['url']
    count+=1
    print('Nama:',nama, 
          'Harga:',harga, 
          'Harga Diskon:',harga_diskon, 
          'Diskon:',besar_diskon, 
          'Stok:',stok, 
          'Terjual:',terjual, 
          'Kota:',kota, 
          'Provinsi:',provinsi, 
          'Link:',link)
    
    #Save Data
    write = csv.writer(open('{}.csv'.format(key), 'a', newline=''))
    data = [nama, 
            harga, 
            harga_diskon, 
            besar_diskon, 
            stok, 
            terjual, 
            kota, 
            provinsi, 
            link]
    write.writerow(data)
```
